### Bonus #2

Filter out the non-single-states (e.g., `'California, Colorado'`) in the most elegant way possible (i.e., *not* by blacklisting all the bad values).

```python

dfSalary['salary'] = pd.to_numeric(dfSalary['salary'], errors='coerce')

df_salary_cleaned = dfSalary.dropna(subset=['salary', 'state'])

df_single_state = df_salary_cleaned[~df_salary_cleaned['state'].str.contains(',', na=False)]


salary_stats_by_single_state = df_single_state.groupby('state')['salary'].agg(['min', 'mean', 'max']).reset_index()

salary_stats_by_single_state
```

### Bonus #3

Show the quantiles instead of just minimum, mean, and maximumâ€•say 0%, 5%, 25%, 50%, 75%, 95%, and 100%. Outliers may be deceiving.

Sort by whatever interests youâ€•like say the *50th* percentile.

And throw in a count by state too. It would be interesting to know how many data points contribute to the figures for each state. (*Hint: your nice formatting from Bonus #1 might not work this time around.* ðŸ˜œ)

```python
dfSalary['salary'] = pd.to_numeric(dfSalary['salary'], errors='coerce')

df_salary_cleaned = dfSalary.dropna(subset=['salary', 'state'])

df_single_state = df_salary_cleaned[~df_salary_cleaned['state'].str.contains(',', na=False)]

quantiles = [0, 0.05, 0.25, 0.5, 0.75, 0.95, 1]

salary_stats_with_quantiles = df_single_state.groupby('state')['salary'].quantile(quantiles).unstack() #The unstack() function reshapes the data, making each quantile a separate column.
salary_stats_with_quantiles['count'] = df_single_state.groupby('state')['salary'].count()

salary_stats_sorted = salary_stats_with_quantiles.sort_values(0.5, ascending=False)


salary_stats_sorted
```
